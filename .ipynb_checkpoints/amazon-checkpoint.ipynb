```python
import boto3
import csv
import re
import os
region='us-east-1'
with open("rekognition.csv","r") as input:
    next(input)
    reader=csv.reader(input)
    for line in reader:
        accesskey=line[2]
        secretkey=line[3]
#photo='coarse_tilt_aligned_face.2282.11597935265_29bcdfa4a5_o.jpg'
client=boto3.client('rekognition',aws_access_key_id=accesskey,aws_secret_access_key=secretkey,region_name=region)
#client=boto3.client('rekognition')


basepath = 'C:/Users/sshar/Desktop/microsoft/faces'
for entry in os.listdir(basepath):
    for en in os.listdir(os.path.join(basepath,entry)):
        x = re.search(".*jpg", en)
        if(x):
            y=os.path.join(basepath,entry)
            z=os.path.join(y,en)
            with open(z, "rb") as source_image:
                image=source_image.read()
                response=client.detect_faces(Image={'Bytes':image}, Attributes=['ALL'])
                for key,value in response.items():
                    if key=='FaceDetails':
                        for k in value:
                            for ke,val in k.items():
                                if ke=='Gender':
                                    print(val)
```
