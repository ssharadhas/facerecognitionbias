```python
#brightness
import sys
from PIL import Image,ImageStat

def brightness( im_file ):
   im = Image.open(im_file).convert('L')
   stat = ImageStat.Stat(im)
   return stat.mean[0]

b=brightness('dataset/yaya.jpg')
print(b)
```

```python
#Skin detection
import cv2
import numpy as np
import matplotlib.pyplot as plt

min_YCrCb = np.array([0,133,77],np.uint8)
max_YCrCb = np.array([235,173,127],np.uint8)

# Get pointer to video frames from primary device
image = cv2.imread("dataset/messi.jpg")
imageYCrCb = cv2.cvtColor(image,cv2.COLOR_BGR2YCR_CB)
skinRegionYCrCb = cv2.inRange(imageYCrCb,min_YCrCb,max_YCrCb)

skinYCrCb = cv2.bitwise_and(image, image, mask = skinRegionYCrCb)

cv2.imwrite("dataset/messi.png", np.hstack([image,skinYCrCb]))
```

```python
import os
p=os.listdir('C:/Users/sshar/googlevision/dataset')
basepath = 'C:/Users/sshar/googlevision/dataset/'
listof=[]
dict={}
for i in p:
    if os.path.isdir(os.path.join(basepath,i)):
        b=os.path.join(basepath,i)
        j=os.listdir(b):
        y=os.path.join(b,j)
        x=['b0.0','b0.5','b1.0','b1.5','b2.0','b4.0','b7.0','b9.0','b11.0','b15.0','b20.0','b25','b30','b35','b40','b45','b50']
        for pattern in x:
            z=re.search(pattern,y)
            if(z):
                with open(y, "rb") as source_image:
                    image=source_image.read()
                    response=client.detect_faces(Image={'Bytes':image}, Attributes=['ALL'])
                    for key,value in response.items():
                        if key=='FaceDetails':
                            for k in value:
                                for ke,val in k.items():
                                    if ke=='Gender':
                                        for key1,val1 in val.items():
                                            if key1=='Confidence':
                                                dict[pattern]=val1;
        listof.append(dict)
```
